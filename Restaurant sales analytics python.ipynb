```python
df = df.dropna(subset = ["sale_id"])
```

```python
#what are the top selling products?
```

```python
top_products = df.groupby("product_name") ["total_price"].sum().sort_values(ascending = False).reset_index().head(5)
```

```python
top_products
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.barplot(data = top_products, x = "product_name", y = "total_price")
plt.title("Top selling products")
plt.ylabel("Total Revenue")
plt.xlabel("Product Name")
```

```python
type(df["sale_date"])
```

```python
df["sale_date"] = pd.to_datetime(df["sale_date"], format = "%Y-%m-%d", errors = "coerce")
```

```python
df.drop('date', axis = 1, inplace = True)
```

```python
df['month'] = pd.to_datetime(df['sale_date']).dt.month_name()
```

```python
# monthly sales trend
```

```python
monthly_trend = df.groupby("month") ["total_price"].sum().reset_index()
month_order = ["January", "February", "November", "December"]
monthly_trend["month"] = pd.Categorical(monthly_trend["month"], categories = month_order, ordered = True)
monthly_trend = monthly_trend.sort_values("month")
```

```python
monthly_trend
```

```python
df['month'].unique()
```

```python
df['month'].nunique()
```

```python
sns.lineplot(data = monthly_trend, x = "month", y = "total_price")
plt.title("Month Sales Trend")
plt.ylabel("Total Revenue")
plt.xlabel("Month Name")
```

```python
# what are the top purchasing locations
```

```python
top_locations = df.groupby("location") ["total_price"].sum().sort_values(ascending = False).reset_index().head(5)
```

```python
top_locations
```

```python
sns.barplot(data = top_locations, x = "total_price", y = "location")
plt.title("Highest Selling Locations")
plt.xlabel("total_price")
plt.ylabel("location")
```

```python
# what is the top purchasing gender(s)?
```

```python
top_gender = df.groupby("gender") ["total_price"].sum().sort_values(ascending = False).reset_index().head()
```

```python
top_gender
```

```python
sns.barplot(data = top_gender, x = "gender", y = "total_price")
plt.title("gender hierarchy in purchase")
plt.xlabel("Genders")
plt.ylabel("Total Price")
plt.ticklabel_format(style = 'plain', axis = 'y')
```
